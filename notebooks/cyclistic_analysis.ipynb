```python
import pandas as pd

# Load each month
jan = pd.read_csv("202501-divvy-tripdata.csv")
feb = pd.read_csv("202502-divvy-tripdata.csv")
mar = pd.read_csv("202503-divvy-tripdata.csv")
apr = pd.read_csv("202504-divvy-tripdata.csv")
may = pd.read_csv("202505-divvy-tripdata.csv")
jun = pd.read_csv("202506-divvy-tripdata.csv")
jul = pd.read_csv("202507-divvy-tripdata.csv")
aug = pd.read_csv("202508-divvy-tripdata.csv")
sep = pd.read_csv("202509-divvy-tripdata.csv")
oct = pd.read_csv("202510-divvy-tripdata.csv")
nov = pd.read_csv("202511-divvy-tripdata.csv")
dec = pd.read_csv("202512-divvy-tripdata.csv")

# combine all files
all_trips = pd.concat(
    [jan, feb, mar, apr, may, jun, jul, aug, sep, oct, nov, dec],
    ignore_index=True
)

# check
print(all_trips.shape)

# convert dates
all_trips['started_at'] = pd.to_datetime(all_trips['started_at'])
all_trips['ended_at'] = pd.to_datetime(all_trips['ended_at'])

# calculate ride length and convert to minutes
all_trips['ride_length'] = (
all_trips['ended_at'] - all_trips['started_at']
).dt.total_seconds()
all_trips['ride_length_minutes'] = all_trips['ride_length'] / 60

# remove null rides
all_trips = all_trips[all_trips['ride_length_minutes'] > 0]

# add day of week
all_trips['day_of_week'] = all_trips['started_at'].dt.day_name()

### Analysis ###
# average ride length
all_trips.groupby('member_casual')['ride_length_minutes'].mean()

# rides by day
all_trips.groupby(['member_casual', 'day_of_week'])['ride_id'].count()

# summary

summary = all_trips.groupby(['member_casual', 'day_of_week']).agg(
    number_of_rides=('ride_id', 'count'),
    avg_ride_length_minutes=('ride_length_minutes', 'mean')
).reset_index()

print(summary)

summary.to_csv("summary.csv", index=False)
```
